```python
#Manage data for variable 1
df['H4GH1'].value_counts()
```

```python
df['health_status']=df['H4GH1']
df['health_status'].value_counts()
```

```python
df['health_status']=df['health_status'].loc[df['health_status']!=' ']
df['health_status'].value_counts()
```

```python
df['health_status'].replace(['1','2','3','4','5'], ['excellent','very good','good','fair','poor'],inplace=True)
df['health_status'].value_counts()
```

```python
rp.summary_cat(df['health_status'])
```

```python
#Manage data for variable 1
df['H4WS4'].value_counts()
```

```python
df['close_friends'] = df['H4WS4']
df['close_friends'].value_counts()
```

```python
df['close_friends'] = df['close_friends'].loc[ df['close_friends'] !=' ']
df['close_friends'].value_counts()
```

```python
df['close_friends']=df['close_friends'].loc[df['close_friends']!='95']
df['close_friends']=df['close_friends'].loc[df['close_friends']!='96']
df['close_friends']=df['close_friends'].loc[df['close_friends']!='98']
df['close_friends'].value_counts()
```

```python
df['close_friends'].replace(['1','2','3','4','5'],['none','1 or 2 friends','3 to 5 friends','6 to 9 friends','10 or more friends'],inplace=True)
df['close_friends'].value_counts()
```

```python
rp.summary_cat(df['close_friends'])
```

```python
#Graphs and tables or statistics
df['close_friends'].value_counts().plot(kind ='pie')
```

```python
df['close_friends'].value_counts(normalize = True).plot(kind ='bar')
```
