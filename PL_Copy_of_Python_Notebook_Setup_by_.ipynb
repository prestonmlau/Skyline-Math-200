```python
#Manage data for variable 1
df['H4MI1'].value_counts()

```

```python
df['serve_military']=df['H4MI1']
df['serve_military'].value_counts()
```

```python
df['serve_military']=df['serve_military'].loc[df['serve_military']!=' ']
df['serve_military'].value_counts()
```

```python
df['serve_military'].replace(['0','1'], ['no','yes',],inplace=True)
df['serve_military'].value_counts()
```

```python
#Graphs and tables or statistics
df['serve_military'].value_counts(normalize=True).plot(kind='pie')
```

```python
df['serve_military'].value_counts(normalize=True).plot(kind='bar')
```

```python
#Manage data for variable 1
df['H4SP5'].value_counts()
```

```python
df['trouble_sleeping'] = df['H4SP5']
df['trouble_sleeping'].value_counts()
```

```python
df['trouble_sleeping'] = df['trouble_sleeping'].loc[ df['trouble_sleeping'] !=' ']
df['trouble_sleeping'].value_counts()
```

```python
df['trouble_sleeping']=df['trouble_sleeping'].loc[df['trouble_sleeping']!='5']
df['trouble_sleeping']=df['trouble_sleeping'].loc[df['trouble_sleeping']!='8']
df['trouble_sleeping'].value_counts()
```

```python
df['trouble_sleeping'].replace(['0','1','2','3','4'],['never in the past four weeks','less than once a week','1 or 2 times a week','3 or 4 times a week','5 or more times a week'],inplace=True)
df['trouble_sleeping'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
df['trouble_sleeping'].value_counts().plot(kind ='pie')
```

```python
df['trouble_sleeping'].value_counts(normalize = True).plot(kind ='bar')
```
